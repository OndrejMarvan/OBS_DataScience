**Objective:** This project explores Unsupervised Learning through Dimensionality Reduction. I start with a high-dimensional dataset (5000 * 5000 distance matrix) and aim to project it into a 2D plane.

**The "Wau" Effect (considered as a challenge):** The algorithm is "blind"—it is only given the distances between cities, never their actual coordinates. Because MDS coordinates are arbitrary, I apply Procrustes Analysis (using Singular Value Decomposition) to mathematically rotate and align the discovered structure with the real-world orientation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from math import radians, cos, sin, asin, sqrt
import time

# 1. DATA PREPARATION
# ---------------------------------------------------------
print("Step 1: Loading 'worldcities.csv' and sampling top cities...") # useful feature to be sure what is being processed right now. 
df = pd.read_csv('worldcities.csv')

# Use of 5,000 cities to ensure the continents self-form from the data density but will run on my machine. Note, total dataset has 48,060 rows (unique cities).
n_cities = 5000 
df_filtered = df.sort_values('population', ascending=False).head(n_cities).reset_index(drop=True)

real_lat = df_filtered['lat'].values
real_lng = df_filtered['lng'].values

# 2. DISTANCE MATRIX CALCULATION (Haversine Formula) - source: https://en.wikipedia.org/wiki/Haversine_formula
# ---------------------------------------------------------
def haversine(lat1, lon1, lat2, lon2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    return 6371 * c # Radius of the Earth. 

print(f"Step 2: Calculating {n_cities**2:,} distances...")
start_time = time.time() 

dist_matrix = np.zeros((n_cities, n_cities))
for i in range(n_cities):
    for j in range(n_cities):
        dist_matrix[i, j] = haversine(real_lat[i], real_lng[i], real_lat[j], real_lng[j])

print(f"   Calculation finished in {round(time.time() - start_time)} seconds.") # To calculate execution time (approx. 42 sec for 5000 cities)
```

## Dimensionality Reduction (MDS)
I apply the MDS algorithm. It treats the distances as "constraints" and tries to find a 2D configuration of points that satisfies those distances as closely as possible.

```python
# 3. RUN MDS ALGORITHM
# ---------------------------------------------------------
print(f"Step 3: Running MDS on {n_cities} points...")
# n_jobs=-1 uses all available CPU cores
mds = MDS(n_components=2, dissimilarity="precomputed", random_state=42, n_init=1, n_jobs=-1)
coords = mds.fit_transform(dist_matrix)
print("   Dimensionality reduction complete.")
```

## Procrustes Alignment (The Rotation Fix)
MDS coordinates are relative, meaning the map could be upside down or mirrored. I use Procrustes Analysis to find the optimal rotation matrix $R$. I solve the "Orthogonal Procrustes Problem" using SVD to align the discovery with the ground truth.

```python
# 4. ALIGNMENT LOGIC
# ---------------------------------------------------------
print("Step 4: Aligning orientation to match global geography...")

# Center both datasets (0-centering)
real_coords = np.column_stack((real_lng, real_lat))
real_centered = real_coords - np.mean(real_coords, axis=0)
mds_centered = coords - np.mean(coords, axis=0)

# Singular Value Decomposition (SVD) for rotation
H = mds_centered.T @ real_centered
U, S, Vt = np.linalg.svd(H)
Rotation = U @ Vt
aligned_coords = mds_centered @ Rotation

# Scale discovery to match the degree scale of Earth
x_final = aligned_coords[:, 0]
y_final = aligned_coords[:, 1]
scale = np.std(real_lng) / np.std(x_final)
x_final = (x_final * scale) + np.mean(real_lng)
y_final = (y_final * scale) + np.mean(real_lat)
```

# Final Visualization
In this final view, I remove all background maps and labels. The "shape" of the world you see is formed entirely by the density of the cities placed by the MDS algorithm.

```python
# 5. FINAL VISUALIZATION
# ---------------------------------------------------------
plt.figure(figsize=(20, 10))

# Plotting the MDS result alone with very small, dense dots
plt.scatter(x_final, y_final, c='crimson', s=1.2, alpha=0.6)

plt.title(f"Reconstructed World Map: {n_cities} Points (Unsupervised MDS)", fontsize=22)
plt.xlim(-180, 180)
plt.ylim(-90, 90)

# Formatting for a clean 'Satellite' look
plt.axis('off') 
plt.gca().set_facecolor('white')

plt.show()
```

**Conclusion**: In this project, I am trying to prove that it's possible to discover the shape of the Earth using nothing but a table of distances. I am starting with a list of 5,000 major cities, but I am hiding their coordinates from the computer. The only information the algorithm gets is a "distance matrix"—a giant grid of numbers showing how far every city is from every other city. This is a classic test of Unsupervised Learning because the computer has to find a hidden 2D structure in a 5,000-dimensional dataset.

To do this, I am using a technique called Multidimensional Scaling (MDS). Think of MDS like a person trying to draw a map by only looking at the mileage signs on a highway. It doesn't know where North or East is, but it knows that if London is 5,500 km from New York, they must be placed that far apart on its drawing. By the time it finishes placing all 5,000 cities, the outlines of the continents should naturally start to appear.

**Further conclusion note**: While geographic coordinates are inherently 2D, this project demonstrates MDS can recover structure from only pairwise dissimilarities - proving the algorithm works as described in course when distances faithfully represent relationships.
