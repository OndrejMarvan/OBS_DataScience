# 1. How to use packages in Python?

The dot (.) means ‚Äúlook inside this thing‚Äù

Think of it like an address:

* Neighborhood ‚Üí package (e.g., pandas)

* House ‚Üí module or object inside the package (e.g., DataFrame, io)

* Room ‚Üí a function, class, or attribute (e.g., read_csv)

Python reads left to right: find the thing on the left, then get the attribute on the right.

### 1) Modules & packages (library functions)
When you import a library, you get a namespace (a ‚Äúbox of names‚Äù). The dot accesses names inside it.

* math, pandas, numpy ‚Üí modules/packages (namespaces)

* sqrt, read_csv, arange ‚Üí functions inside those namespaces

````python
import math
math.sqrt(9)          # use sqrt inside the math module

import pandas as pd
pd.read_csv("sample_people.csv")  # use read_csv inside pandas (aliased as pd)

import numpy as np
np.arange(5)          # use arange inside numpy (aliased as np)
````

### 2) Objects & methods (functions attached to data)

The dot also calls methods (functions that belong to an object):

````python
text = "hello"
text.upper()       # method on a string object
````

````python
df = pd.DataFrame({"x": [1, 2, 3]})
df.head(2)         # method on a DataFrame object
````

### 3) Chaining with dots
You can chain dots when you keep looking inside things:

````python
pd.DataFrame({"x":[1,2,3]})
````

````python
pd.DataFrame({"x":[1,2,3]}).x.mean()
# DataFrame(...) -> column x (a Series) -> its method mean()
````

# Checking is something is a certain type

### Python Type Checking ‚Äî `isinstance()` and `type()`

## 1. Why Check Types?

Sometimes you want your program to behave differently depending on what kind of data you‚Äôre working with ‚Äî  
for example, numbers vs strings, or lists vs dictionaries.

Python provides **two main tools** for that:

- `isinstance()` ‚Äî the *recommended*, flexible way  
- `type()` ‚Äî a direct, strict way

---

## 2. `isinstance()`

### Basic syntax
```python
isinstance(object, type)

```
---
### 3. `type()`

#### Basic syntax
```python
type(object)

````python
a = 5
b = "hello"

print(isinstance(a, int))       # True
print(isinstance(b, str))       # True
print(isinstance(a, float))     # False
print(isinstance(a, (int, float)))  # True (checks multiple types)
````

````python
x = 3.14
print(type(x))         # <class 'float'>
print(type("hi"))      # <class 'str'>
````

### Comparing `isinstance()` vs `type()`

| **Feature** | **`isinstance()`** | **`type()`** |
|--------------|--------------------|---------------|
| Checks subclasses | ‚úÖ Yes | ‚ùå No |
| Can check multiple types | ‚úÖ Yes, using tuple | ‚ö†Ô∏è No |
| Returns | `True` / `False` | Type object (`<class 'int'>`) |
| Recommended for | Most type checks | Exact type match only |


### Summary

* Use isinstance() when checking categories of types (e.g. ‚Äúis this a number?‚Äù).
* Use type() only when you need an exact match.
* Always prefer isinstance() in real-world code ‚Äî it‚Äôs more flexible and Pythonic.

# More complex objects

**`list`** ‚Äî ordered, changeable collection.

### What is a list?
A **list** is an ordered, changeable (mutable) collection that can hold **mixed types**.
Think of it like a flexible box that keeps items in order and lets you add, remove, or replace them.

Key properties

* Ordered: items keep their position ([1, 2, 3] is different from [3, 2, 1]).
* Mutable: you can change items in place.
* Heterogeneous: different types can live together (ints, strings, lists, ‚Ä¶).
* Indexable & sliceable: access by position with [].


````python
nums = [1, 2, 3]
print(nums)
nums.append(4)     # [1, 2, 3, 4]
nums[0]            # 1
print(nums)
````

````python
len(nums)
````

````python
nested = [
    [1, 2, 3],      # row 0
    [4, 5, 6],      # row 1
    [7, 8, 9],      # row 2
]

nested[0]        # [1, 2, 3]       (first inner list)
nested[0][1]     # 2               (row 0, column 1)
nested[-1][-1]   # 9               (last row, last column)
````

````python
nested[1][2] = 99     # change value in row 1, col 2
# nested is now [[1,2,3], [4,5,99], [7,8,9]]
nested
````

#### Changing lists (mutation)

````python
nums = [1, 2, 3]
print(nums)
# Replace
nums[0] = 10            # [10, 2, 3]

# Add
nums.append(4)          # [10, 2, 3, 4]    ‚Üê add one at the end
nums.extend([5, 6])     # [10, 2, 3, 4, 5, 6]  ‚Üê add many
nums.insert(1, 99)      # [10, 99, 2, 3, 4, 5, 6]  ‚Üê insert at position

# Remove
nums.remove(99)         # removes first matching value ‚Üí [10, 2, 3, 4, 5, 6]
last = nums.pop()       # removes & returns last item (6)
second = nums.pop(1)    # removes & returns item at index 1 (2)
del nums[0]             # delete by index ‚Üí [3, 4, 5]

print(nums)
````

#### Searching & membership

````python
vals = [1, 2, 2, 3]
2 in vals        # True
vals.count(2)    # 2  (how many times 2 appears)
vals.index(3)    # 3's first index ‚Üí 3
````

## Why do we need ‚ÄúTypes of Objects in Python‚Äù?

Knowing the main object types isn‚Äôt trivia‚Äîit tells you **what you can do** with a value, **how Python stores it**, and **which tool fits the job**. Types drive:
- **Operations:** `3 + 2` works, `"3" + "2"` concatenates, `3 + "2"` errors.
- **Behavior & methods:** strings have `.upper()`, lists have `.append()`, dicts have `.keys()`.
- **Performance & memory:** numbers vs arrays vs DataFrames have very different speed/size trade-offs.
- **APIs & libraries:** pandas expects columns as specific dtypes; NumPy arrays must be numeric for fast math.
- **Correctness:** choosing the right type avoids bugs (e.g., `None` vs empty string vs `0`).

---

### Mental model: ‚Äúpick the right container for the job‚Äù
| Goal | Best fit (usually) | Why |
|---|---|---|
| Single number/text/flag | `int`, `float`, `str`, `bool` | Simple scalars with the right operations |
| Ordered collection you can edit | `list` | Keep order, allow duplicates, mutate easily |
| Fixed small record | `tuple` | Immutable, lightweight |
| Unique membership / fast ‚Äúin?‚Äù | `set` | No duplicates, fast lookup |
| Labeled key‚Üívalue data | `dict` | Natural for records / JSON-like data |
| Big numeric grids | **NumPy `ndarray`** | Vectorized math, fast storage |
| Labeled column/ table data | **pandas `Series` / `DataFrame`** | Column names, indices, groupby/merge/pivot |

````python
# My Solution - Lists
# 1. Create nums and append 1
nums = [3, 1, 4]
nums.append(1)

# 2. Insert 9 at the beginning
nums.insert(0, 9)

# 3. Replace the middle element of [10, 20, 30]
mid_replace = [10, 20, 30]
mid_replace[1] = 25

# 4. Slicing letters
letters = ["a", "b", "c", "d", "e"]
sliced = letters[1:4]

# 5. Concatenate lists
combined = [1, 2] + [3, 4]

# 6. Shallow copy
original = [1, [2, 3]]
shallow = original.copy()

# 7. Sorted copy (original data remains [5, 2, 9, 1])
data = [5, 2, 9, 1]
data_sorted = sorted(data)

# V√Ωstupy pro kontrolu
print(f"1 & 2 (nums): {nums}")
print(f"3 (middle):  {mid_replace}")
print(f"4 (slice):   {sliced}")
print(f"5 (concat):  {combined}")
print(f"6 (shallow): {shallow} (is original: {shallow is original})")
print(f"7 (sorted):  {data_sorted}, original: {data}")
````

````python
# My Solution - Tuples

# 1. Create a 3-item tuple and access its last item
triplet = ("SK", "CZ", "PL")
last_item = triplet[-1]

# 2. Convert tuple -> list -> change -> back to tuple
letters_tuple = ("x", "y", "z")
temp_list = list(letters_tuple)  # Konverze na mƒõniteln√Ω list
temp_list[1] = "Y"               # Zmƒõna "y" na "Y"
final_tuple = tuple(temp_list)   # Konverze zpƒõt na nemƒõnn√Ω tuple

# V√Ωstupy pro kontrolu
print(f"1. Cel√Ω tuple: {triplet}, Posledn√≠ prvek: {last_item}")
print(f"2. P≈Øvodn√≠: {letters_tuple} -> Upraven√Ω: {final_tuple}")
````

````python
# 1. Strip and lower (Method Chaining)
s = "  Hello, World!  "
cleaned = s.strip().lower()

# 2. String to list (Split)
names_str = "Ana,Bob,Chi"
names_list = names_str.split(",")

# 3. List to string (Join)
chars = ["a", "b", "c"]
joined_str = "-".join(chars)

# 4. Slicing documentation
word = "documentation"
first_four = word[:4]
last_three = word[-3:]

# V√Ωstupy pro kontrolu
print(f"1. Vyƒçi≈°tƒõno: '{cleaned}'")
print(f"2. Seznam:    {names_list}")
print(f"3. Spojeno:   '{joined_str}'")
print(f"4. Slicing:   Prvn√≠ 4: '{first_four}', Posledn√≠ 3: '{last_three}'")
````

# 3. What is pandas?
**pandas** is a Python library for working with **tabular data**. It provides two core data structures:

- `Series` ‚Äî a 1D labeled array (think: a single column)
- `DataFrame` ‚Äî a 2D labeled table (think: a spreadsheet)

We'll create a tiny dataset to explore the basics.


1. **Creating DataFrames** ‚Äî Building tabular data in pandas (e.g., from dicts, lists, CSV/Excel/SQL) into a `DataFrame`.

2. **Subset Observations (rows)** ‚Äî Selecting specific rows by labels/positions or conditions (e.g., `df.loc[...]`, `df[df.col > 0]`).

3. **Subset Variables (columns)** ‚Äî Selecting one or more columns by name (e.g., `df["age"]`, `df[["name","age"]]`).

4. **Subsets (rows and columns)** ‚Äî Selecting both at once with label/position indexers (e.g., `df.loc[row_sel, col_sel]`, `df.iloc[...]`).

5. **Make New Columns** ‚Äî Creating or transforming columns via vectorized expressions, string ops, or `np.where`/`assign`.

Next time ;)

6. **Reshaping Data** ‚Äî Changing table layout: sort rows/columns, rename labels, reindex to new labels, pivot/widen, melt/longen.

7. **Combine Data Sets** ‚Äî Merging, joining, or concatenating multiple tables (e.g., `merge`, `join`, `concat`).

8. **Group Data** ‚Äî Splitting into groups and aggregating or transforming (e.g., `groupby(...).agg(...)`, `transform`).

9. **Handling Missing Data** ‚Äî Detecting, filling, or dropping nulls (e.g., `isna`, `fillna`, `dropna`).

10. **Summarize Data** ‚Äî Getting descriptive stats and counts (e.g., `describe`, `value_counts`, `agg`, `mean/median/sum`).

11. **Plotting** ‚Äî Quick visualizations from DataFrames/Series (e.g., `df.plot(...)`) or with Matplotlib/Seaborn.

12. **Changing Type** ‚Äî Converting dtypes for columns (e.g., `astype`, `to_numeric`, `to_datetime`, `Categorical`).

13. **Datetime** ‚Äî Working with dates/times: parsing, extracting parts, resampling, and time-based indexing (`to_datetime`, `.dt`, `date_range`).


### Examples: pandas `Series` (1D) and `DataFrame` (2D)

`Series` ‚Äî a 1D labeled array (a single column)

````python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)
````

### Access by label or position:

````python
s["b"]     # 20
s.iloc[0]  # 10
````

### Vectorized operations:

````python
s * 2      # a 20, b 40, c 60
(s >= 20)  # a False, b True, c True
````

### Useful attributes/methods:

````python
s.index          # Index(['a','b','c'], dtype='object')
s.values         # array([10, 20, 30])
s.mean()         # 20.0
s.rename("scores")
````

`DataFrame` ‚Äî a 2D labeled table (like a spreadsheet)

````python
df = pd.DataFrame({
    "name": ["Ana", "Bob", "Chi"],
    "age":  [28,    17,   35],
    "city": ["London", "Paris", "Berlin"]
})
print(df)
#    name  age    city
# 0   Ana   28  London
# 1   Bob   17   Paris
# 2   Chi   35  Berlin
````

````python
df["age"]                # a Series (one column)
df[["name", "city"]]     # a DataFrame (two columns)

df.loc[1, "name"]        # label-based ‚Üí 'Bob'
df.iloc[0, 2]            # position-based ‚Üí 'London'
````

# Summary

* Series = one labeled column (1D).

* DataFrame = table with labeled rows & columns (2D).

* Both support vectorized operations (fast, no loops needed).


## Vectorized operations ‚Äî explained for a beginner

**Idea:** Instead of looping through items one-by-one, you tell Python (via NumPy/pandas)  
to do the operation on the **whole column/array at once**. It‚Äôs shorter, clearer, and faster.

---

### Why it‚Äôs useful
- **Less code:** one line instead of many.
- **Faster:** heavy lifting happens in optimized C code under the hood.
- **Fewer bugs:** no index mistakes, no manual counters.

---

## With NumPy arrays

**Without vectorization (loopy style):**
```python
# (For illustration) add 10 to each element
data = [1, 2, 3, 4]
out = []
for x in data:
    out.append(x + 10)
# out = [11, 12, 13, 14]

Perform the following operations on the DataFrame:

2. Display all flowers that are **not** of the species `setosa`.  
3. Find rows where `sepal_length` is between **5 and 6** (inclusive).  
4. Display only rows where `sepal_length > 6` **and** `petal_width > 1.5`.


## 3. Creating and Sorting Columns

1. Create a new column named `sepal_area`, which is the result of multiplying `sepal_length * sepal_width`.  
2. Sort the DataFrame **in descending order** by the `petal_length` column.  
3. Display only the **first 5 rows** where `sepal_area > 20`.

---


![Iris Dataset](https://gadictos.com/wp-content/uploads/2019/03/iris-machinelearning.png)


````python
import seaborn as sns
import pandas as pd

# 1. LOAD THE DATA
# Naƒçten√≠ datasetu Iris
iris = sns.load_dataset('iris')

print("--- 1. Z√ÅKLADN√ç INFO ---")
print(f"Prvn√≠ch 10 ≈ô√°dk≈Ø:\n{iris.head(10)}")
print("-" * 30)
iris.info()
print("-" * 30)
print(f"Shape (tvar): {iris.shape}")


# 2. FILTERING DATA
# 2.1 Petal length > 4.5
filt_length = iris[iris['petal_length'] > 4.5]

# 2.2 Nen√≠ 'setosa'
filt_not_setosa = iris[iris['species'] != 'setosa']

# 2.3 Sepal length mezi 5 a 6 (vƒçetnƒõ)
filt_between = iris[iris['sepal_length'].between(5, 6)]

# 2.4 Sepal_length > 6 AND petal_width > 1.5
# Nutno pou≈æ√≠t z√°vorky a oper√°tor &
filt_and = iris[(iris['sepal_length'] > 6) & (iris['petal_width'] > 1.5)]


# 3. CREATING AND SORTING
# 3.1 Nov√Ω sloupec sepal_area
iris['sepal_area'] = iris['sepal_length'] * iris['sepal_width']

# 3.2 Se≈ôazen√≠ podle petal_length sestupnƒõ
iris_sorted = iris.sort_values(by='petal_length', ascending=False)

# 3.3 Prvn√≠ch 5 ≈ô√°dk≈Ø, kde sepal_area > 20
result_area = iris[iris['sepal_area'] > 20].head(5)


# V√ùSTUPY PRO KONTROLU
print("\n--- 2.4 FILTR (Length > 6 & Width > 1.5) ---")
print(filt_and.head())

print("\n--- 3.3 V√ùSLEDEK (Area > 20, prvn√≠ch 5) ---")
print(result_area)
````
