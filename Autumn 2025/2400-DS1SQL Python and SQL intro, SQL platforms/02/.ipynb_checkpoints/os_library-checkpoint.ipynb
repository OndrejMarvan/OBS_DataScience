## 1Ô∏è‚É£ What is the `os` module?

The **`os`** module gives you access to many **operating system functions** directly from Python ‚Äî  
such as working with **files, directories, and environment variables**.

To use it:

```python
import os


````python
import os

# Get current working directory (where Python is running)
print(os.getcwd())
````

````python
print(os.listdir())  # list everything in the current folder
````

# You can also specify a path:

````python

print(os.listdir("/Users/ewaweychert/Desktop"))   # on Linux/Mac
````

````python
os.mkdir("test_folder")      # create a folder
print(os.listdir())          # verify it‚Äôs there

os.rmdir("test_folder")      # remove the folder
print(os.listdir())          # verify it‚Äôs gone
````

### üóÇÔ∏è Common `os` Module Functions

| **Task** | **Function** |
|-----------|--------------|
| Get current directory | `os.getcwd()` |
| Change directory | `os.chdir(path)` |
| List files/folders | `os.listdir(path)` |
| Create folder | `os.mkdir(path)` |
| Remove folder | `os.rmdir(path)` |
| Join paths | `os.path.join(a, b)` |
| Check if file/folder exists | `os.path.exists(path)` |


### üì¶ Where Packages Are Installed

| **Where you run the command** | **Where the package goes** | **Jupyter may use it?** |
|-------------------------------|-----------------------------|--------------------------|
| `pip install ...` in system terminal | System Python site-packages | ‚ùå Not always |
| Inside Jupyter Notebook (`!pip install ...`) | Kernel‚Äôs environment | ‚úÖ Yes |

---

### ‚öôÔ∏è Common Problems and Fixes

| **Problem** | **Cause** | **Solution** |
|--------------|------------|---------------|
| `ModuleNotFoundError` | Library installed in system Python, not kernel env | Use `!pip install` inside notebook |
| Wrong interpreter | Kernel points to a different Python | Check `sys.executable` |
| Conda env mismatch | Notebook uses base env | Register env with `ipykernel install` |
| Global vs virtual env | Different `site-packages` directories | Ensure kernel and pip target same environment |
