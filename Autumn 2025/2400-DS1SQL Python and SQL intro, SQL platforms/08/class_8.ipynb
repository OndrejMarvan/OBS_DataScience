<p><strong>What is object-oriented programming (OOP)?</strong></p>
<p><strong>Definition:</strong> Object-Oriented Programming (OOP) is a programming paradigm that organizes code based on objects. These objects contain both data (attributes) and behaviors (methods), allowing for clearer and more logical code structures.</p>

<p><strong>Why do we need OOP instead of just functions?</strong></p>
<p>The traditional procedural approach relies on functions and variables, but as code becomes more complex, managing it becomes more difficult. Object-oriented programming solves this by grouping data and operations on that data within objects, which improves code organization and readability.</p>

<p>Key advantages of OOP:</p>
<ul>
<li><strong>Code reusability –</strong> The ability to reuse classes and objects in different parts of a program or in other projects.</li>
<li><strong>Modularity –</strong> Code is divided into independent modules (classes), making it easier to manage and develop.</li>
<li><strong>Scalability –</strong> OOP makes it easier to extend applications without major changes to existing code.</li>
<li><strong>Easier debugging –</strong> Thanks to the division into classes and objects, it is easier to track errors and diagnose issues in the code.</li>
</ul>

<p>Ultimately, OOP helps write more organized, flexible, and maintainable code.</p>


<p>Object-oriented programming</p>

<p>Object-oriented programming (OOP) is a technique that helps manage complex software projects. Most analysts will not work independently on such solutions, but this knowledge is very useful, for example when creating your own package.</p>

<p>The concept of object-oriented programming allows you to look holistically at the process of collecting and processing data. It involves defining new types of objects (called <strong>classes</strong>) and the functions or operations assigned to them (called <strong>methods</strong>).</p>

<p><strong>Objects</strong> are, in this approach, concrete instances of classes, while classes are a way of grouping objects into sets of units with shared characteristics. For each class we can define a list of characteristics, and then assign specific values of those characteristics to each object.</p>

<p>Working with classes allows us to more efficiently organize our knowledge and create representations of real-world objects within computer programs.</p>

<p>For example, salmon, eel, and whale shark are representatives of the class “fish” (even though they differ greatly from each other). If we say “Nemo is a fish,” then even people who don’t know the Disney movie can easily list some characteristics of that “object”:</p>

<p>* it has fins, <br>
* it breathes with gills <br>
* it has no voice.</p>

<p>Most programming languages have a single class system.</p>

<p>In R, four independent class systems are available: S3, S4, Reference Class (RC or R5), and the increasingly popular R6 system. We also have the “base” system, which contains the most fundamental classes of objects in R: <code>numeric</code>, <code>character</code>, <code>factor</code>, <code>list</code>, etc.</p>

<p>These systems coexist, each with different features and properties, which we will discuss below.</p>

<p>Programming language features that are important from the perspective of object-oriented programming:</p>

<p>* polymorphism — the ability to create <strong>methods</strong> (functions) whose behavior (result) may depend on the type (<strong>class</strong>) of the object passed as an argument<br>
* inheritance — the ability to create a new class based on an existing one by specifying that the new class is a special case of the existing class.</p>

<p>Attributes of the datetime.datetime class<br>

<ul>
<li>now() → returns the current date and time</li>
<li>today() → returns today’s date (without time)</li>

<p>Summary<br>

It has attributes (e.g. year, month, day).<br>

````python
from datetime import datetime

# Creating a date and time object
now = datetime.now()
print(now)

# Attributes of the datetime.datetime class
print(now.year)   # Year
print(now.month)  # Month
print(now.day)    # Day
print(now.hour)   # Hour
print(now.minute) # Minute
print(now.second) # Second

# Methods of the datetime.datetime class
print(now.strftime("%Y-%m-%d %H:%M:%S"))  # Date formatting
print(now.weekday())  # Day of the week (0 = Monday, 6 = Sunday)

print(now.strftime("%d-%m-%Y %H:%M"))  # Format: DD-MM-YYYY HH:MM
````

````python
class Student:
    def __init__(self, first_name, last_name, student_id, year_of_study, grades, graduated):
        """Initializes a Student object"""
        self.first_name = first_name
        self.last_name = last_name
        self.student_id = student_id
        self.year_of_study = year_of_study
        self.grades = grades
        self.graduated = graduated

    def average(self):
        """Calculates the student's average grade"""
        if self.grades:
            return sum(self.grades) / len(self.grades)
        return 0  # If no grades, return 0

    def info(self):
        """Returns basic information about the student"""
        return f"{self.first_name} {self.last_name} ({self.student_id})"


# Creating Student objects
s1 = Student("Piotr", "Wiśniewski", 123456, 3, [2, 3, 4, 5, 4.5], False)
s2 = Student("Maria", "Niewiadoma", 987654, 4, [3.5, 5, 5, 4, 3, 4, 5, 4.5], False)

# Tests:
print(s1.average())      # Result: 3.7
print(s2.average())      # Result: 4.25

print(s1.info())         # Result: "Piotr Wiśniewski (123456)"
print(s2.info())         # Result: "Maria Niewiadoma (987654)"
````

<h3><strong>What should happen?</strong></h3>

<ol>
  <li>
    <p><strong>Correct cases:</strong></p>
    <ul>
      <li>If the data is valid, a <code>Student</code> object will be created and the methods will work correctly.</li>
      <li><code>average()</code> will calculate the average of the grades while ignoring <code>None</code> values.</li>
      <li><code>info()</code> will return the student’s first name, last name, and student ID number.</li>
    </ul>
  </li>

  <li>
    <p><strong>Incorrect cases (ValueError errors):</strong></p>
    <ul>
      <li>If <code>student_id</code> <strong>does not have exactly 6 digits</strong> → an error should be raised.</li>
      <li>If <code>year_of_study</code> <strong>is not a number between 1 and 5</strong> → an error should be raised.</li>
      <li>If <code>grades</code> <strong>do not contain exactly 10 elements</strong> or <strong>contain invalid values</strong> → an error should be raised.</li>
    </ul>
  </li>
</ol>


````python
class Student:
    ALLOWED_GRADES = {2, 2.5, 3, 3.5, 4, 4.5, 5, None}  # Allowed grade values

    def __init__(self, first_name, last_name, student_id, year_of_study, grades, graduated):
        """Initialize the Student object with data validation."""

        self.first_name = first_name
        self.last_name = last_name
        self.graduated = graduated

        # Validate student ID (must have exactly 6 digits)
        if isinstance(student_id, int) and 100000 <= student_id <= 999999:
            self.student_id = student_id
        else:
            raise ValueError("Student ID must be a six-digit number!")

        # Validate year of study (must be between 1 and 5)
        if isinstance(year_of_study, int) and 1 <= year_of_study <= 5:
            self.year_of_study = year_of_study
        else:
            raise ValueError("Year of study must be an integer between 1 and 5!")

        # Validate grades (must contain exactly 10 values and all must be allowed)
        if (
            isinstance(grades, list)
            and len(grades) == 10
            and all(g in self.ALLOWED_GRADES for g in grades)
        ):
            self.grades = grades
        else:
            raise ValueError(
                "Grades must be a list of 10 elements containing only: "
                "2, 2.5, 3, 3.5, 4, 4.5, 5, or None!"
            )

    def average(self):
        """Calculates the student's average grade, ignoring missing grades (None)."""
        clean_grades = [g for g in self.grades if g is not None]
        return sum(clean_grades) / len(clean_grades) if clean_grades else 0

    def info(self):
        """Returns basic information about the student."""
        return f"{self.first_name} {self.last_name} ({self.student_id})"


# Tests:

try:
    s1 = Student("Piotr", "Wiśniewski", 123456, 3,
                 [2, 3, 4, 5, 4.5, 3, 2.5, 5, 4, None], False)
    print(s1.info())    # "Piotr Wiśniewski (123456)"
    print(s1.average()) # Calculated average

    s2 = Student("Maria", "Niewiadoma", 987654, 4,
                 [3.5, 5, 5, 4, 3, 4, 5, 4.5, None, None], False)
    print(s2.info())    # "Maria Niewiadoma (987654)"
    print(s2.average()) # Calculated average

    # Test with incorrect data (should raise an exception)
    s3 = Student("Jan", "Kowalski", 12345, 2,       # Error: invalid student ID
                 [3, 3, 4, 5, 3.5, 4, 5, 4.5, 5, 5], True)

except ValueError as e:
    print(f"Error: {e}")
````

<h2><strong>Example 1: Implementing a Class for Cleaning Data in a Pandas DataFrame</strong></h2>

<p>Your task is to implement a <code>DataCleaner</code> class that allows cleaning and processing data stored in a Pandas DataFrame.  
The class should include a set of methods that enable removing duplicates, filling missing values, and normalizing column names.</p>

<p><strong>Requirements</strong></p>

<p><strong>Class Initialization</strong></p>
<p>The class should take a Pandas DataFrame as an argument and store its copy so that the original data remains unchanged.</p>

<p><strong>Method: remove_duplicates()</strong></p>
<p>This method should remove all duplicated rows in the DataFrame and return the modified class object.</p>

<p><strong>Method: fill_missing(strategy, columns)</strong></p>
<p>This method should fill missing values according to the selected strategy:</p>

<ul>
  <li>"mean" – fills missing values with the mean</li>
  <li>"median" – fills missing values with the median</li>
  <li>"mode" – fills missing values with the mode (most common value)</li>
  <li>or any constant value provided by the user</li>
</ul>

<p>The method should work only on numerical columns.<br>
If the list of columns is not provided, it should fill missing values in all numerical columns.</p>

<p><strong>Method: normalize_column_names()</strong></p>
<p>This method should convert all column names to lowercase and replace spaces with underscores (_).</p>

<p><strong>Method: get_cleaned_data()</strong></p>
<p>This method should return the processed and cleaned DataFrame.</p>


````python
import seaborn as sns
print(sns.get_dataset_names())  
df_cars = sns.load_dataset("mpg")
df_cars.head()

df_1 = df_cars.copy()

df_1 = df_cars
````

````python
import pandas as pd

class DataCleaner:
    """
    A class for cleaning data inside a Pandas DataFrame.
    It contains various useful data-cleaning methods.
    """

    def __init__(self, df):
        """
        Initializes a DataCleaner object based on the provided DataFrame.
        Creates a copy to avoid modifying the original dataset.
        """
        self.df = df.copy()

    def remove_duplicates(self):
        """
        Removes duplicated rows from the dataset.

        Returns:
        - self (for method chaining)
        """
        self.df = self.df.drop_duplicates()
        return self  # Return the object, not df

    def fill_missing(self, strategy="mean", columns=None):
        """
        Fills missing values in the dataset using the selected strategy.

        Parameters:
        - strategy: "mean", "median", "mode", or a constant value
        - columns: list of columns to process (all if None)

        Returns:
        - self (for method chaining)
        """
        if columns is None:
            columns = self.df.columns

        for col in columns:
            if self.df[col].dtype in ["int64", "float64"]:
                if strategy == "mean":
                    self.df[col] = self.df[col].fillna(self.df[col].mean())
                elif strategy == "median":
                    self.df[col] = self.df[col].fillna(self.df[col].median())
                elif strategy == "mode":
                    self.df[col] = self.df[col].fillna(self.df[col].mode()[0])
                else:
                    self.df[col] = self.df[col].fillna(strategy)

        return self  # Return the object

    def normalize_column_names(self):
        """
        Standardizes column names by making them lowercase 
        and replacing spaces with underscores.

        Returns:
        - self (for method chaining)
        """
        self.df.columns = self.df.columns.str.lower().str.replace(" ", "_")
        return self

    def get_cleaned_data(self):
        """
        Returns the cleaned Pandas DataFrame.
        """
        return self.df


data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Eve", None],
    "Age": [25, 30, 35, 25, None, 40],
    "Salary": [50000, 60000, None, 50000, 75000, 80000],
    "Department": ["HR", "IT", "IT", "HR", "Finance", "HR"]
}
data = pd.DataFrame(data)

cleaner = DataCleaner(data)

cleaned_df_class = (
    cleaner
        .remove_duplicates()
        .fill_missing(strategy="median", columns=["Age", "Salary"])
        .normalize_column_names()
        .get_cleaned_data()
)

cleaned_df_class
````

        <strong>Filling missing values</strong> in the <code>"Age"</code> and <code>"Salary"</code>  
        columns using the <code>"median"</code> strategy,
      </li>
      <li><strong>Normalizing the column names</strong>.</li>
    </ul>
  </li>
</ul>


````python
import pandas as pd

def remove_duplicates(df):
    """
    Removes duplicated rows from the dataset.
    
    Parameters:
    - df: Pandas DataFrame

    Returns:
    - DataFrame without duplicates.
    """
    return df.drop_duplicates()

def fill_missing(df, strategy="mean", columns=None):
    """
    Fills missing values in the dataset using the selected strategy.

    Parameters:
    - df: Pandas DataFrame
    - strategy: Method for filling missing values:
        - "mean"
        - "median"
        - "mode"
        - or any constant value provided by the user
    - columns: List of columns in which missing values should be filled. 
               If None, the method applies to all columns.

    Returns:
    - DataFrame with missing values filled.
    """
    df = df.copy()  # Create a copy to avoid modifying the original DataFrame

    if columns is None:
        columns = df.columns  # If no columns specified, use all columns

    for col in columns:
        if df[col].dtype in ["int64", "float64"]:  # Process only numeric columns
            if strategy == "mean":
                df[col] = df[col].fillna(df[col].mean())  # Fill with mean
            elif strategy == "median":
                df[col] = df[col].fillna(df[col].median())  # Fill with median
            elif strategy == "mode":
                df[col] = df[col].fillna(df[col].mode()[0])  # Fill with mode (most frequent value)
            else:
                df[col] = df[col].fillna(strategy)  # Fill with a constant value
                
    return df

def normalize_column_names(df):
    """
    Normalizes column names by converting them to lowercase 
    and replacing spaces with underscores "_".
    
    Parameters:
    - df: Pandas DataFrame

    Returns:
    - DataFrame with standardized column names.
    """
    df = df.copy()  # Create a copy to avoid modifying the original DataFrame
    df.columns = df.columns.str.lower().str.replace(" ", "_")
    return df

# Example data
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Eve", None],
    "Age": [25, 30, 35, 25, None, 40],
    "Salary": [50000, 60000, None, 50000, 75000, 80000],
    "Department": ["HR", "IT", "IT", "HR", "Finance", "HR"]
}
data = pd.DataFrame(data)

# Calling functions sequentially (without using pipe)
cleaned_df_functions = remove_duplicates(data)
cleaned_df_functions = fill_missing(cleaned_df_functions, strategy="median", columns=["Age", "Salary"])
cleaned_df_functions = normalize_column_names(cleaned_df_functions)

cleaned_df_functions
````
