### What Are Univariate Plots?
Univariate plots show the distribution of a single variable.
#### The variable can be:
* categorical (e.g., race, gender),
* quantitative (e.g., age, weight).
The goal is to show how often particular values or ranges occur.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load data
df = pd.read_csv("Marriage.csv")

# Count occurrences in the 'race' column
race_counts = df['race'].value_counts()
races = race_counts.index
counts = race_counts.values

# Create folder (if it doesn't already exist)
save_folder = "plots"
os.makedirs(save_folder, exist_ok=True)

# Create the plot
plt.figure(figsize=(8, 5))
bars = plt.bar(races, counts, color="skyblue", edgecolor="black")
plt.title("Distribution of Race Categories (Marriage)", fontsize=14, fontweight='bold')
plt.xlabel("Race", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{int(height)}',
             ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# ðŸ”½ Save as PNG in the folder
save_path = os.path.join(save_folder, "race_distribution.png")
plt.savefig(save_path, dpi=300, bbox_inches='tight')  # dpi=300 â†’ high quality
plt.show()
plt.close()  # closes the plot (prevents window pop-up)

print(f"Plot saved as: {save_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load data
df = pd.read_csv("Marriage.csv")

# Create folder for saving outputs
save_folder = "plots"
os.makedirs(save_folder, exist_ok=True)

# Prepare data â€” frequency table
pivot_df = pd.crosstab(df['person'], df['race'])

# Normalize to proportions (each bar = 100%)
pivot_prop = pivot_df.div(pivot_df.sum(axis=1), axis=0)

# Create 100% stacked bar chart
fig, ax = plt.subplots(figsize=(8, 5))
pivot_prop.plot(
    kind='bar',
    stacked=True,
    color=plt.cm.Set2.colors,  # pastel palette
    edgecolor='black',
    ax=ax
)

# Title and axis labels
plt.title("Race Share by Person (Marriage)", fontsize=14, fontweight='bold')
plt.xlabel("Person", fontsize=12)
plt.ylabel("Share (%)", fontsize=12)

# Add percentage labels to segments
for container in ax.containers:
    labels = [f"{v.get_height()*100:.1f}%" if v.get_height() > 0 else "" for v in container]
    ax.bar_label(container, labels=labels, label_type='center', fontsize=8, color='black')

# Legend
plt.legend(title="Race", loc='upper right', bbox_to_anchor=(1.25, 1))

# Grid and styling
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.ylim(0, 1)  # ensure y-axis ends at 100%
plt.tight_layout()

# Save plot
save_path = os.path.join(save_folder, "stacked_bar_race_by_person_100pct.png")
plt.savefig(save_path, dpi=300, bbox_inches='tight')

plt.show()
plt.close()

print(f"Plot saved as: {save_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from matplotlib.ticker import StrMethodFormatter

# --- 1) Load data ------------------------------------------------------------
# Expected file: Salaries.csv (columns include: 'rank', 'salary')
df = pd.read_csv("Salaries.csv")

# --- 2) Aggregation: mean salary by rank -------------------------------------
plotdata = (
    df.groupby("rank", as_index=False)["salary"]
      .mean()
      .rename(columns={"salary": "mean_salary"})
)

# Optional: enforce rank order
order = ["AsstProf", "AssocProf", "Prof"]
plotdata["rank"] = pd.Categorical(plotdata["rank"], categories=order, ordered=True)
plotdata = plotdata.sort_values("rank")

# --- 3) Create bar chart ------------------------------------------------------
plt.figure(figsize=(7.5, 5))
bars = plt.bar(
    plotdata["rank"],
    plotdata["mean_salary"],
    color="cornflowerblue",
    edgecolor="black"
)

# Title and axis labels
plt.title("Average Salary by Academic Rank (9-month salary, 2008â€“2009)",
          fontsize=13, fontweight="bold")
plt.xlabel("")
plt.ylabel("Average Salary")

# Format Y-axis as currency (e.g., $120,000)
plt.gca().yaxis.set_major_formatter(StrMethodFormatter("${x:,.0f}"))

# Add value labels on bars (currency format)
for bar, val in zip(bars, plotdata["mean_salary"]):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() * 1.01,
        f"${val:,.2f}",
        ha="center", va="bottom", fontsize=10
    )

# X-axis labels with nicer names
labels_map = {
    "AsstProf": "Assistant\nProfessor",
    "AssocProf": "Associate\nProfessor",
    "Prof": "Full\nProfessor"
}
plt.gca().set_xticklabels([labels_map.get(x.get_text(), x.get_text()) for x in plt.gca().get_xticklabels()])

# Grid and layout
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()

# --- 4) Save to file ---------------------------------------------------------
os.makedirs("plots", exist_ok=True)
save_path = os.path.join("plots", "mean_salary_by_rank.png")
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.show()
plt.close()

print("Data used for the plot:\n", plotdata)
print(f"Plot saved as: {save_path}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from scipy import stats as st

# --- 0) Load data -------------------------------------------------------------
# The file should contain at least the columns: 'rank', 'sex', 'salary'
df = pd.read_csv("Salaries.csv")

# (optional) order the ranks
rank_order = ["AsstProf", "AssocProf", "Prof"]
if set(rank_order).issubset(set(df['rank'].unique())):
    df['rank'] = pd.Categorical(df['rank'], categories=rank_order, ordered=True)

# Folder for saving plots (optional)
save_folder = "plots"
os.makedirs(save_folder, exist_ok=True)

# --- 1) Basic statistics by rank ---------------------------------------------
g = df.groupby('rank', observed=False)['salary']
plotdata_rank = g.agg(n='count', mean='mean', sd='std').reset_index()
plotdata_rank['se'] = plotdata_rank['sd'] / np.sqrt(plotdata_rank['n'])
# 95% CI: t * SE
plotdata_rank['tcrit'] = plotdata_rank['n'].apply(
    lambda n: st.t.ppf(0.975, n-1) if n > 1 else np.nan
)
plotdata_rank['ci'] = plotdata_rank['tcrit'] * plotdata_rank['se']

# --- 2) Plot: Mean + SE by rank ----------------------------------------------
x = np.arange(len(plotdata_rank))  # positions on the X-axis
fig, ax = plt.subplots(figsize=(8, 5))
ax.errorbar(
    x,
    plotdata_rank['mean'],
    yerr=plotdata_rank['se'],
    fmt='o',
    capsize=4,
    elinewidth=1.2,
    color='black',
    label='Mean Â± SE'
)
ax.plot(x, plotdata_rank['mean'], linestyle='-', color='black', alpha=0.7)

ax.set_title("Mean salary by rank (Â±SE)", fontsize=14, fontweight='bold')
ax.set_xlabel("Rank", fontsize=12)
ax.set_ylabel("Mean salary", fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(plotdata_rank['rank'])
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# (optional) save
path1 = os.path.join(save_folder, "mean_se_by_rank.png")
plt.savefig(path1, dpi=300, bbox_inches='tight')
plt.show()
plt.close()
print(f"Saved: {path1}")

# --- 3) Statistics by rank and sex -------------------------------------------
g2 = df.groupby(['rank', 'sex'], observed=False)['salary']
plotdata_sex = g2.agg(n='count', mean='mean', sd='std').reset_index()
plotdata_sex['se'] = plotdata_sex['sd'] / np.sqrt(plotdata_sex['n'])
plotdata_sex['tcrit'] = plotdata_sex['n'].apply(
    lambda n: st.t.ppf(0.975, n-1) if n > 1 else np.nan
)
plotdata_sex['ci'] = plotdata_sex['tcrit'] * plotdata_sex['se']

# Ensure rank order on the X-axis
if 'rank' in plotdata_sex and isinstance(df['rank'].dtype, pd.CategoricalDtype):
    plotdata_sex['rank'] = plotdata_sex['rank'].astype(
        pd.CategoricalDtype(df['rank'].cat.categories, ordered=True)
    )
plotdata_sex = plotdata_sex.sort_values(['sex', 'rank'])

# --- 4) Plot: Mean + SE by rank and sex (overlaid) ---------------------------
fig, ax = plt.subplots(figsize=(8, 5))
for sex, sub in plotdata_sex.groupby('sex'):
    # X positions for this sex (consistent with rank order)
    ranks = sub['rank'].astype(str).tolist()
    x = np.arange(len(ranks))
    ax.errorbar(
        x,
        sub['mean'],
        yerr=sub['se'],
        fmt='o',
        capsize=4,
        elinewidth=1.2,
        label=str(sex)
    )
    ax.plot(x, sub['mean'], linestyle='-', alpha=0.8)

ax.set_title("Mean salary by rank and sex (Â±SE)", fontsize=14, fontweight='bold')
ax.set_xlabel("Rank", fontsize=12)
ax.set_ylabel("Mean salary", fontsize=12)
ax.set_xticks(np.arange(len(rank_order)))
ax.set_xticklabels(rank_order)
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.legend(title="Sex")
plt.tight_layout()

path2 = os.path.join(save_folder, "mean_se_by_rank_and_sex_overlap.png")
plt.savefig(path2, dpi=300, bbox_inches='tight')
plt.show()
plt.close()
print(f"Saved: {path2}")

# --- 5) Plot: â€œdodgedâ€ (side-by-side by sex) ---------------------------------
# Map ranks to numeric positions
rank_to_pos = {r: i for i, r in enumerate(rank_order)}
plotdata_sex['x'] = plotdata_sex['rank'].astype(str).map(rank_to_pos)

# Determine order of sexes for horizontal offsets
sex_levels = list(plotdata_sex['sex'].dropna().unique())
sex_levels.sort()
w = 0.35  # half-width of offset
offsets = np.linspace(-w, w, num=len(sex_levels)) if len(sex_levels) > 1 else [0.0]

fig, ax = plt.subplots(figsize=(8, 5))
for off, sex in zip(offsets, sex_levels):
    sub = plotdata_sex[plotdata_sex['sex'] == sex].sort_values('rank')
    x = sub['x'].values + off
    ax.errorbar(
        x,
        sub['mean'],
        yerr=sub['se'],
        fmt='o',
        capsize=4,
        elinewidth=1.2,
        label=str(sex)
    )
    ax.plot(x, sub['mean'], linestyle='-', alpha=0.8)

ax.set_title("Mean salary by rank and sex (Â±SE) â€” dodged", fontsize=14, fontweight='bold')
ax.set_xlabel("Rank", fontsize=12)
ax.set_ylabel("Mean salary", fontsize=12)
ax.set_xticks(np.arange(len(rank_order)))
ax.set_xticklabels(rank_order)
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.legend(title="Sex")
plt.tight_layout()

path3 = os.path.join(save_folder, "mean_se_by_rank_and_sex_dodged.png")
plt.savefig(path3, dpi=300, bbox_inches='tight')
plt.show()
plt.close()
print(f"Saved: {path3}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

# --- 1) Load data --------------------------------------------------------------
df = pd.read_csv("Salaries.csv")  # columns: salary, sex, rank, discipline

# Order categories
rank_order = ["AsstProf", "AssocProf", "Prof"]
disc_map   = {"A": "Theoretical", "B": "Applied"}
sex_order  = ["Male", "Female"]

df["rank"] = pd.Categorical(df["rank"], categories=rank_order, ordered=True)
df["discipline"] = pd.Categorical(
    df["discipline"].map(disc_map),
    categories=["Theoretical", "Applied"],
    ordered=True
)
df["sex"] = pd.Categorical(df["sex"], categories=sex_order, ordered=True)

# --- 2) Aggregation: mean, SD, SE ---------------------------------------------
agg = (
    df.groupby(["rank", "discipline", "sex"], observed=True)
      .agg(
          n=("salary", "size"),
          mean=("salary", "mean"),
          sd=("salary", "std")
      )
      .reset_index()
)
agg["se"] = agg["sd"] / np.sqrt(agg["n"])

# --- 3) Helper function: draw mean with SE in each facet ----------------------
sns.set_theme(style="whitegrid")
palette = sns.color_palette("Set1", n_colors=2)  # colors: Male / Female

def draw_points_with_se(data, **kws):
    ax = plt.gca()
    order = ["Male", "Female"]
    x_positions = np.linspace(-0.01, 0.01, len(order))  # points close to each other
    
    for i, sex in enumerate(order):
        subset = data[data["sex"] == sex]
        if not subset.empty:
            ax.scatter(x_positions[i], subset["mean"], s=50, c=[palette[i]], zorder=3)
            ax.errorbar(
                x_positions[i],
                subset["mean"],
                yerr=subset["se"],
                fmt="none",
                ecolor="black",
                elinewidth=1,
                capsize=3,
                zorder=2
            )

    # X-axis centered (no labels)
    ax.set_xticks([0])
    ax.set_xticklabels([""], fontsize=8)

# --- 4) FacetGrid --------------------------------------------------------------
g = sns.FacetGrid(
    agg,
    col="rank", col_order=rank_order,
    row="discipline", row_order=["Theoretical", "Applied"],
    sharey=True, margin_titles=True,
    height=3.0, aspect=1.1
)
g.map_dataframe(draw_points_with_se)

# --- 5) Y-axis formatting and titles ------------------------------------------
ymin = max(0, (agg["mean"] - 2 * agg["se"]).min())
ymax = (agg["mean"] + 2 * agg["se"]).max()
for ax in g.axes.flat:
    ax.set_ylim(ymin, ymax)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f"${v:,.0f}"))

g.set_axis_labels("", "")
g.set_titles(row_template="{row_name}", col_template="{col_name}")
g.fig.suptitle(
    "Nine-month academic salaries by sex, discipline, and rank",
    fontsize=14,
    fontweight="bold"
)
g.fig.subplots_adjust(top=0.88, hspace=0.25, wspace=0.15)

# --- 6) Global legend ----------------------------------------------------------
handles = [
    plt.Line2D([0], [0], marker='o', color='w', label='Male',
               markerfacecolor=palette[0], markersize=8),
    plt.Line2D([0], [0], marker='o', color='w', label='Female',
               markerfacecolor=palette[1], markersize=8)
]
g.fig.legend(
    handles=handles,
    title="Sex",
    loc='lower center',
    ncol=2,
    bbox_to_anchor=(0.5, -0.02)
)

# --- 7) Save -------------------------------------------------------------------
os.makedirs("plots", exist_ok=True)
outpath = os.path.join("plots", "mean_se_by_sex_facets_rank_discipline.png")
plt.savefig(outpath, dpi=300, bbox_inches="tight")
plt.show()
plt.close()

print(f"âœ… Saved: {outpath}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 1) Data from gapminder
from gapminder import gapminder

# --- filtering: Americas + years 1952 and 2007 ---
plotdata_long = (
    gapminder.loc[
        (gapminder["continent"] == "Americas") &
        (gapminder["year"].isin([1952, 2007])),
        ["country", "year", "lifeExp"]
    ]
)

# 2) Wide table: columns y1952 and y2007
plotdata_wide = (
    plotdata_long
    .pivot(index="country", columns="year", values="lifeExp")
    .rename(columns={1952: "y1952", 2007: "y2007"})
    .reset_index()
)

# Sort by the 1952 value (ascending) to make the plot readable
plotdata_wide = plotdata_wide.sort_values(by="y1952", ascending=True)

# 3) Drawing the dumbbell chart
fig, ax = plt.subplots(figsize=(12, 9), facecolor="white")

# Background lines (horizontal â€œrailsâ€)
ax.hlines(
    y=plotdata_wide["country"],
    xmin=plotdata_wide["y1952"],
    xmax=plotdata_wide["y2007"],
    color="#758D99",
    alpha=0.6,
    linewidth=2,
    zorder=1
)

# Points for 1952 and 2007
ax.scatter(
    plotdata_wide["y1952"], plotdata_wide["country"],
    s=60, color="#DB444B", label="1952", zorder=3
)
ax.scatter(
    plotdata_wide["y2007"], plotdata_wide["country"],
    s=60, color="#006BA2", label="2007", zorder=3
)

# X and Y axis
ax.set_xlabel("Life Expectancy (years)", fontsize=12)
ax.set_yticks(plotdata_wide["country"])
ax.set_yticklabels(plotdata_wide["country"], ha="left")
ax.yaxis.set_tick_params(pad=110, labelsize=10)  # larger offset for country labels

# Grid and style
ax.grid(which="major", axis="x", linestyle="--", alpha=0.4)
ax.set_axisbelow(True)

# Legend and title
ax.legend(loc="lower right", frameon=False)
ax.set_title("Change in Life Expectancy â€” Americas (1952 â†’ 2007)", fontsize=14, fontweight="bold")

plt.tight_layout()

# 4) Save to file
os.makedirs("plots", exist_ok=True)
outpath = os.path.join("plots", "dumbbell_americas_1952_2007.png")
plt.savefig(outpath, dpi=300, bbox_inches="tight")
plt.show()
plt.close()

print(f"Saved: {outpath}")
```

```python
import pandas as pd
import os
import plotly.graph_objects as go
import kaleido

# --- 1) Load data ------------------------------------------------------------
df = pd.read_csv("Marriage.csv")

# Keep only required columns and remove empty rows
df = df[['person', 'race']].dropna()

# --- 2) Count flows person -> race ------------------------------------------
flows = (
    df.groupby(['person', 'race'], dropna=False)
      .size()
      .reset_index(name='value')
)

# --- 3) Nodes and indices ----------------------------------------------------
persons = sorted(flows['person'].unique().tolist())
races   = sorted(flows['race'].unique().tolist())

nodes = persons + races
idx = {name: i for i, name in enumerate(nodes)}

source = [idx[p] for p in flows['person']]
target = [idx[r] for r in flows['race']]
value  = flows['value'].tolist()

# --- 4) Colors ---------------------------------------------------------------
person_colors = ["#1f77b4", "#ff7f0e"]  # Bride, Groom
race_colors   = ["#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf",
                 "#d62728", "#1f9e89", "#636efa", "#ef553b", "#00cc96"]

node_colors = []
for n in nodes:
    if n in persons:
        node_colors.append(person_colors[persons.index(n) % len(person_colors)])
    else:
        node_colors.append(race_colors[races.index(n) % len(race_colors)])

def rgba(hex_color, alpha=0.5):
    hex_color = hex_color.lstrip('#')
    r = int(hex_color[0:2], 16)
    g = int(hex_color[2:4], 16)
    b = int(hex_color[4:6], 16)
    return f"rgba({r},{g},{b},{alpha})"

link_colors = [rgba(node_colors[t], 0.35) for t in target]

# --- 5) Create the figure ----------------------------------------------------
fig = go.Figure(data=[go.Sankey(
    arrangement="snap",
    node=dict(
        pad=18,
        thickness=18,
        line=dict(color="black", width=0.5),
        label=nodes,
        color=node_colors
    ),
    link=dict(
        source=source,
        target=target,
        value=value,
        color=link_colors
    )
)])

fig.update_layout(
    title_text="Flow: Person (Bride/Groom) â†’ Race (Marriage.csv)",
    font=dict(size=12),
    margin=dict(l=10, r=10, t=60, b=10),
    width=900, height=550
)

# # --- 6) Save PNG + preview ---------------------------------------------------
# os.makedirs("plots", exist_ok=True)
# png_path = os.path.join("plots", "sankey_marriage.png")

# # to save PNG, kaleido is required â†’ pip install -U kaleido
# fig.write_image(png_path, scale=2)
# fig.show()

# print(f"Plot saved as: {png_path}")
```

```python
# --- Heatmap: visualization of relationships between numerical variables ---
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# --- 1) Load data ------------------------------------------------------------
df = pd.read_csv("Salaries.csv")

# Select only numerical columns
num_df = df.select_dtypes(include="number")

# --- 2) Standardization (Z-score) -------------------------------------------
scaled_df = (num_df - num_df.mean()) / num_df.std(ddof=0)

# --- 3) Create folder for saving --------------------------------------------
save_folder = "plots"
os.makedirs(save_folder, exist_ok=True)

# --- 4) Draw heatmap ---------------------------------------------------------
plt.figure(figsize=(10, 5))
sns.heatmap(
    scaled_df.corr(),              # correlation matrix
    annot=True,                    # values inside cells
    cmap="coolwarm",               # color palette
    fmt=".2f",                     # number formatting
    linewidths=0.5,                # lines between cells
    cbar_kws={"label": "Correlation"}
)

plt.title("Correlation Matrix â€” Salaries.csv", fontsize=14, fontweight='bold')
plt.tight_layout()

# --- 5) Save and show --------------------------------------------------------
save_path = os.path.join(save_folder, "heatmap_salaries_corr.png")
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.show()
plt.close()

print(f"Plot saved as: {save_path}")
```

```python
# --- 1) Imports ---------------------------------------------------------------
# !pip install wordcloud
import os
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# --- 2) Data: we will use an example dataset from seaborn --------------------
import seaborn as sns
df = sns.load_dataset("mpg").dropna()

# Combine text from the "origin", "name", and "model_year" columns as the word source
text = " ".join(df["name"].astype(str) + " " + df["origin"].astype(str) + " " + df["model_year"].astype(str))

# --- 3) Create word cloud ----------------------------------------------------
stopwords = set(STOPWORDS)
wordcloud = WordCloud(
    background_color="white",
    width=800,
    height=400,
    max_words=100,
    colormap="viridis",
    stopwords=stopwords
).generate(text)

# --- 4) Visualization ---------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud â€“ Car Names and Origins (mpg dataset)", fontsize=14, fontweight='bold')
plt.tight_layout()

# --- 5) Save to file ----------------------------------------------------------
os.makedirs("plots", exist_ok=True)
save_path = os.path.join("plots", "wordcloud_mpg.png")
wordcloud.to_file(save_path)

plt.show()
plt.close()

print(f"âœ… Plot saved as: {save_path}")
```
