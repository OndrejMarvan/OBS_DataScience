<h2 data-start="63" data-end="92">Computational Complexity</h2>
<p data-start="94" data-end="357"><strong data-start="94" data-end="109">Definition:</strong><br data-start="109" data-end="112" />A way to describe how the <strong data-start="138" data-end="156">execution time</strong> or <strong data-start="160" data-end="176">memory usage</strong> of a program grows depending on the size of the input data.<br data-start="236" data-end="239" />It is expressed using <strong data-start="261" data-end="283">Big-O notation (O)</strong>, which represents the approximate rate of growth of computational cost.</p>
<hr data-start="359" data-end="362" />
<h3 data-start="364" data-end="403">1. <strong data-start="371" data-end="401">O(1) &ndash; Constant Complexity</strong></h3>
<p data-start="404" data-end="529">The execution time does <strong data-start="428" data-end="442">not depend</strong> on the size of the input data.<br data-start="473" data-end="476" /><strong data-start="476" data-end="488">Example:</strong> Accessing an array element by its index.</p>
<hr data-start="531" data-end="534" />
<h3 data-start="536" data-end="582">2. <strong data-start="543" data-end="580">O(log n) &ndash; Logarithmic Complexity</strong></h3>
<p data-start="583" data-end="707">The value of <em data-start="596" data-end="599">n</em> is reduced by half in each iteration.<br data-start="637" data-end="640" />Typical for <strong data-start="652" data-end="674">divide and conquer</strong> algorithms, e.g., binary search.</p>
<hr data-start="709" data-end="712" />
<h3 data-start="714" data-end="751">3. <strong data-start="721" data-end="749">O(n) &ndash; Linear Complexity</strong></h3>
<p data-start="752" data-end="870">The execution time grows <strong data-start="777" data-end="795">proportionally</strong> to the size of the data.<br data-start="820" data-end="823" /><strong data-start="823" data-end="835">Example:</strong> Iterating through an entire array.</p>
<hr data-start="872" data-end="875" />
<h3 data-start="877" data-end="924">4. <strong data-start="884" data-end="922">O(n log n) &ndash; Log-Linear Complexity</strong></h3>
<p data-start="925" data-end="1001">Common in <strong data-start="935" data-end="967">efficient sorting algorithms</strong>, such as merge sort or quicksort.</p>
<hr data-start="1003" data-end="1006" />
<h3 data-start="1008" data-end="1049">5. <strong data-start="1015" data-end="1047">O(n&sup2;) &ndash; Quadratic Complexity</strong></h3>
<p data-start="1050" data-end="1160">Each element is compared with every other element.<br data-start="1100" data-end="1103" /><strong data-start="1103" data-end="1115">Example:</strong> Bubble Sort or simple matrix multiplication.</p>
<hr data-start="1162" data-end="1165" />
<h3 data-start="1167" data-end="1210">6. <strong data-start="1174" data-end="1208">O(2ⁿ) &ndash; Exponential Complexity</strong></h3>
<p data-start="1211" data-end="1356">The number of operations <strong data-start="1236" data-end="1247">doubles</strong> with each increase in <em data-start="1270" data-end="1273">n</em>.<br data-start="1274" data-end="1277" /><strong data-start="1277" data-end="1289">Example:</strong> Recursive computation of Fibonacci numbers (without optimization).</p>
<hr data-start="1358" data-end="1361" />
<h3 data-start="1363" data-end="1404">7. <strong data-start="1370" data-end="1402">O(n!) &ndash; Factorial Complexity</strong></h3>
<p data-start="1405" data-end="1556">The number of operations grows <strong data-start="1436" data-end="1457">extremely rapidly</strong> with input size.<br data-start="1474" data-end="1477" />Typical for <strong data-start="1489" data-end="1509">NP-hard problems</strong>, e.g., generating all possible permutations.</p>

<h2 style="text-align: center;">Fibonacci - example of ineffective code</h2>

<h3 style="text-align: center;"><strong><img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/ea9dac862682a4bca20322cdf6e62a3924a00732" alt="" width="540" height="116" /></strong></h3

```python
def fibRecursive(n):
    if n <= 1:
        return n
    return fibRecursive(n-1) + fibRecursive(n-2)
```

```python
print(fibRecursive(2))
print(fibRecursive(3))
print(fibRecursive(4))
print(fibRecursive(5))
print(fibRecursive(6))
print(fibRecursive(7))
```

```python
def fibIterative(n):
    if n <= 1:
        return n
    prev2 = 0
    prev1 = 1
    idx = 1
    while idx < n:
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current
        idx += 1
    return prev1
```

```python
print(fibIterative(2))
print(fibIterative(3))
print(fibIterative(4))
print(fibIterative(5))
print(fibIterative(6))
print(fibIterative(7))
```

```python
print(fibIterative(10_000))
```

```python
print(fibRecursive(40))
```

```python
def fibIterativePrint(n):
    if n <= 1:
        return n
    prev2 = 0
    prev1 = 1
    idx = 1
    while idx < n:
        print(f"licze {idx - 1} + {idx}")
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current
        idx += 1
    return prev1
```

```python

print(fibIterativePrint(7))
```

```python
### This function discards the previous result and starts over.
```

```python
def fibRecursivePrint(n):
    if n <= 1:
        return n
    print(f"licze {n- 2} + {n - 1}")
    return fibRecursivePrint(n-1) + fibRecursivePrint(n-2)
```

```python
print(fibRecursivePrint(7))
```

<p data-start="385" data-end="397"><strong data-start="385" data-end="397">Summary:</strong></p>

<li data-start="399" data-end="466">

</li>
<li data-start="467" data-end="498">
<p data-start="469" data-end="498"><strong data-start="469" data-end="477">O(n)</strong> &ndash; linear functions</p>
</li>

<hr data-start="538" data-end="541" />
<p data-start="543" data-end="771">Each recursive call for <em data-start="567" data-end="570">n</em> results in <strong data-start="582" data-end="606">two additional calls</strong> for <em data-start="611" data-end="616">n-1</em> and <em data-start="621" data-end="626">n-2</em>.<br data-start="627" data-end="630" />This means the number of function calls grows <strong data-start="676" data-end="693">exponentially</strong>, because each call produces two more calls, and those calls branch further.</p>
<p data-start="773" data-end="896">The <strong data-start="777" data-end="795">recursion tree</strong> grows exponentially &mdash; its <strong data-start="822" data-end="832">height</strong> is <em data-start="836" data-end="842">O(n)</em>, and the <strong data-start="852" data-end="871">number of nodes</strong> is approximately <em data-start="889" data-end="893">2ⁿ</em>.</p>
<p data-start="898" data-end="1044">Therefore, the <strong data-start="913" data-end="932">time complexity</strong> of this function is <strong data-start="953" data-end="962">O(2ⁿ)</strong>, which indicates a very fast increase in the number of operations as <em data-start="1032" data-end="1035">n</em> grows.</p>
<p data-start="1046" data-end="1270">This implementation is <strong data-start="1069" data-end="1106">inefficient for large values of n</strong>, because it repeatedly recalculates the same values.<br data-start="1159" data-end="1162" />For <em data-start="1166" data-end="1174">n = 40</em>, the number of operations becomes extremely large, causing significant computation slowdowns.</p>
<p data-start="1272" data-end="1408" data-is-last-node="" data-is-only-node="">A better approach is to use <strong data-start="1300" data-end="1323">dynamic programming</strong> (memoization or an iterative version), which reduces the execution time to <strong data-start="1399" data-end="1407">O(n)</strong>.</p>

```python
import timeit
import pandas as pd

# List for storing results
results_list = []

# Run the benchmark for 50 iterations
for i in range(50):
    # Measure the execution time of the first solution (recursive Fibonacci)
    time_first_solution = timeit.timeit(
        stmt='''
print(fibRecursive(20))
        ''',
        setup='''
def fibRecursive(n):
    if n <= 1:
        return n
    return fibRecursive(n-1) + fibRecursive(n-2)
        ''',
        number=1  # how many times to repeat the execution (1)
    )

    # Measure the execution time of the second solution (iterative Fibonacci)
    time_second_solution = timeit.timeit(
        stmt='''
print(fibIterative(20))
        ''',
        setup='''
def fibIterative(n):
    if n <= 1:
        return n
    prev2 = 0
    prev1 = 1
    idx = 1
    while idx < n:
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current
        idx += 1
    return prev1
        ''',
        number=1
    )

    # Add the results as dictionaries
    results_list.append({
        'solution': 'recursive_function',
        'iteration': i,  # record the exact time for each single run
        'time': time_first_solution
    })
    results_list.append({
        'solution': 'while_loop',
        'iteration': i,  # record the exact time for each single run
        'time': time_second_solution
    })

# Convert the results list into a DataFrame
frame = pd.DataFrame(results_list)

# Display the results
frame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the first plot (boxplot)
sns.boxplot(x="solution", y="time", data=frame)
plt.title("Comparison of execution time distribution for two methods")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

Calculate their <strong data-start="179" data-end="187">mean</strong> and <strong data-start="192" data-end="204">variance</strong> using:</p>
<ol data-start="215" data-end="278">

</li>

<p data-start="257" data-end="278"><strong data-start="257" data-end="276">NumPy functions</strong></p>
</li>

<hr data-start="319" data-end="322" />
<p data-start="324" data-end="497" data-is-last-node="" data-is-only-node=""><strong data-start="324" data-end="350">Why is this important?</strong><br data-start="350" data-end="353" />It demonstrates how <strong data-start="373" data-end="403">NumPy optimizes operations</strong> on large datasets, making calculations much faster and more efficient than pure Python loops.</p>

```python
import numpy as np
import time

# Create 10 million random numbers
data = np.random.rand(10_000_000)

# Calculations in pure Python
start = time.time()
mean_py = sum(data) / len(data)
variance_py = sum((x - mean_py) ** 2 for x in data) / len(data)
end = time.time()
print(f"Python: Mean={mean_py:.5f}, Variance={variance_py:.5f}, Time={end - start:.3f}s")

# Calculations using NumPy
start = time.time()
mean_np = np.mean(data)
variance_np = np.var(data)
end = time.time()
print(f"NumPy: Mean={mean_np:.5f}, Variance={variance_np:.5f}, Time={end - start:.3f}s")
```

<h3 data-start="48" data-end="89">Task 2: Detecting Anomalies in Data</h3>
<p data-start="91" data-end="226"><strong data-start="91" data-end="103">Problem:</strong><br data-start="103" data-end="106" />You have an array of numbers and want to find <strong data-start="152" data-end="164">outliers</strong> &mdash; values greater than the <strong data-start="191" data-end="223">mean + 2 standard deviations</strong>.</p>

```python
data = np.random.normal(50, 15, 1000)  # Simulated data (mean=50, std=15)
mean = np.mean(data)
std = np.std(data)

outliers = data[data > mean + 2 * std]  # Find outlier values
print(f"Found {len(outliers)} anomalies: {outliers}")
```

```python
import random
import statistics

# Simulate data: 1000 random numbers from a normal distribution (mean=50, std=15)
data = [random.gauss(50, 15) for _ in range(1000)]

# Calculate mean and standard deviation using the statistics module
mean = statistics.mean(data)
std = statistics.stdev(data)

# Find outlier values (greater than mean + 2 standard deviations)
outliers = [x for x in data if x > mean + 2 * std]
```

```python
import timeit
import pandas as pd

# List for storing results
results_list = []

# Run the benchmark for 50 iterations
for i in range(50):
    # Measure the execution time of the first solution (using the statistics module and a loop)
    time_first_solution = timeit.timeit(
        stmt='''
# Calculate mean and standard deviation using the statistics module
mean = statistics.mean(data)
std = statistics.stdev(data)

# Find outlier values (greater than mean + 2 standard deviations)
outliers = [x for x in data if x > mean + 2 * std]
        ''',
        setup='''
import random
import statistics
# Simulate data: 1000 random numbers from a normal distribution (mean=50, std=15)
data = [random.gauss(50, 15) for _ in range(1000)]
        ''',
        number=1  # how many times to repeat the test (1)
    )

    # Measure the execution time of the second solution (using NumPy)
    time_second_solution = timeit.timeit(
        stmt='''
mean = np.mean(data)
std = np.std(data)
outliers = data[data > mean + 2 * std]  # Find outlier values
        ''',
        setup='''
import numpy as np
# Simulate data: 1000 random numbers from a normal distribution (mean=50, std=15)
data = np.random.normal(50, 15, 1000)
        ''',
        number=1
    )

    # Add results as dictionaries
    results_list.append({
        'solution': 'loop',
        'iteration': i,  # store exact execution time for each run
        'time': time_first_solution
    })
    results_list.append({
        'solution': 'numpy',
        'iteration': i,  # store exact execution time for each run
        'time': time_second_solution
    })

# Convert the results list into a DataFrame
frame = pd.DataFrame(results_list)

# Display the results
frame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the first plot (boxplot)
sns.boxplot(x="solution", y="time", data=frame)
plt.title("Comparison of execution time distribution for two methods")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
