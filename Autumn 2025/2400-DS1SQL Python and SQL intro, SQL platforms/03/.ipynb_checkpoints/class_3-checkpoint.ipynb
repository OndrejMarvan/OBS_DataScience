<h3>Types of Loops</h3>
<p>A loop is one of the fundamental programming constructs. It allows a selected sequence of instructions to be repeated a specific number of times or until certain conditions are met.</p>
<p>We can distinguish two main types of loops:</p>
<ul>
<li>iterative (counter-controlled)</li>
<li>repetitive (condition-controlled)</li>
</ul>
<p>An <strong>iterative loop</strong> (or <strong>counter-controlled loop</strong>) is a type of loop in which a defined number of iterations (repetitions) are executed. A special variable, called the <strong>control variable</strong> (also known as the <strong>loop counter</strong> or <strong>control variable</strong>), is used to manage the loop’s execution. Moving to the next iteration involves changing the value of the control variable by a specified amount and checking whether the new value remains within the defined range. If it does, the loop repeats (another iteration is executed); if it does not, the loop ends and the program proceeds to the next instruction. In Python, this type of loop is defined with the statement `<span class="hljs-keyword">for</span> variable <span class="hljs-keyword">in</span> set_of_values:<span class="hljs-comment"> COMMANDS</span>`.</p>
<p><br /><strong>A repetitive loop</strong> (or <strong>conditional loop</strong>) is a type of loop in which performing the next iteration depends on a condition defined by the programmer. The condition in the loop definition is an expression that returns a logical (boolean) value. Usually, the t

Loops can also be nested within each other. Let's create an example matrix and fill it with values:


```python
import numpy as np

# Creating a 10x3 matrix filled with zeros
matrix = np.zeros((10, 3), dtype=int)
print(matrix)
# Initialize the number
number = 21

print(matrix.shape[0])
# Loop to iterate through rows and columns
for i in range(matrix.shape[0]):  # Iterate through rows
    for j in range(matrix.shape[1]):  # Iterate through columns
        matrix[i, j] = number
        number += 1  # Increase the counter by 1 number = number +1

# Display the resulting matrix
print(matrix)
print(number)
```

<h2>2. Repetitive Loop — the `while()` Statement</h2>
<p>A repetitive loop in R can be executed using the `while()` statement: `while(CONDITION): COMMANDS`.  
This should be understood as: as long as the CONDITION is true, execute the COMMANDS.  
For the `while()` loop to be finite, one of the COMMANDS must affect the CONDITION.  
Otherwise, you will get an infinite loop (provided the condition was initially true).  
In a `while()` loop, the condition is checked BEFORE executing the COMMANDS.  
Therefore, if the condition is false at the moment the loop is called, no iterations will be executed.</p>
<p>For example:</p>


```python
# First loop

# First while loop: while i < 6
# The loop will run as long as the variable i is less than 6.
# The value of i increases by 1 with each iteration.

i = 1

while i < 6:
    print(i)
    i += 1  # Increment i by 1
```

```python
# Second loop

# Second while loop: while i < 6
# This loop will not execute because the initial value of i is already 6,
# so the condition i < 6 is false from the very beginning.

i = 6

while i < 6:
    print(i)
    i += 1  # Increment i by 1
```

```python
# The while True loop in Python is used to create an infinite loop.
# As long as the condition in while is always True, the loop will run indefinitely.

# while True:
#     print("Ewa")

count = 0

while True:
    print("Ewa")
    count += 1
    if count >= 5:
        break  # Exit the loop after 5 iterations
```

```python
x = 4

while True:  # Infinite loop
    print(x)
    x += 1  # Increment x by 1
    if x >= 5:  # Loop break condition
        break
```

1. **while True:** Creates an infinite loop that will run until the `break` statement is used.  
2. **x += 1:** Increases the value of the variable `x` by 1 in each iteration.  
3. **if x >= 5:** Checks whether the value of `x` has reached 5 or more.  
4. **break:** Ends the loop when `x` is greater than or equal to 5.


<h2>3. The `break` Statement</h2>

The `break` statement can be used in both `while` and `for` loops. It always causes the loop to terminate and moves the program execution to the next commands.  
Similarly, the `next` statement can be used to end the current iteration and proceed to the next one.


```python
for i in range(-10, 11):  # Iterate from -10 to 10 (inclusive)
    if -5 <= i <= 5:  # If i is within the range [-5, 5], break the loop
        break
    print(i)
```

### Explanation:
* **range(-10, 11):** Creates a range of numbers from -10 to 10 inclusive.  
* **if -5 <= i <= 5:** Checks whether the value of `i` is within the range from -5 to 5. If so, the loop is terminated using `break`.  
* **print(i):** If the condition is not met, the value of `i` is printed.  
* The loop stops when `i` reaches -5.


<h2>4. The `continue` Statement</h2>


```python
for i in range(-10, 11):  # Iterate from -10 to 10 (inclusive)
    if -5 <= i <= 5:  # If i is within the range [-5, 5], skip to the next iteration
        continue
    print(i)
```

### Explanation:
* **continue:** Instead of stopping the loop, it moves to the next iteration, skipping the execution of `print(i)` for values in the range -5 <= i <= 5.  
* **print(i):** Prints the value of `i` only when it is outside the range [-5, 5].  

We can use both statements in the same loop, and different types of loops can be nested within each other.

In the following example, for numbers 2–10, we will check:

* the power at which the given number first exceeds **1000 (`1e3`)**,  
* the power at which the given number first exceeds **1,000,000 (`1e6`)**.


```python
# for loop iterating through numbers from 2 to 10
for number in range(2, 11):
    power = 1

    # while loop (equivalent to repeat in R)
    while True:
        # Check if the next power exceeds 1000 for the first time
        if number ** power > 1e3 and number ** (power - 1) <= 1e3:
            print(f"The {power}th power of {number} exceeds 1000 for the first time")
            power += 1
            continue  # Continue to the next iteration of the while loop

        # Check if the next power exceeds 1,000,000
        if number ** power > 1e6:
            print(f"The {power}th power of {number} exceeds 1,000,000 for the first time")
            break  # End the while loop

        # Increase the power by 1
        power += 1
```

```python
# Sliding Window (Moving Average)
data = [10, 20, 30, 40, 50]
window_size = 3
moving_avg = []

for i in range(len(data) - window_size + 1):
    window = data[i:i + window_size]
    avg = sum(window) / window_size
    moving_avg.append(avg)

print("Moving average:", moving_avg)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Example dataset ---
df = pd.DataFrame({
    "Age": np.random.randint(18, 65, 100),
    "Salary": np.random.normal(50000, 10000, 100),
    "Department": np.random.choice(["IT", "HR", "Finance", "Marketing"], 100),
    "Hire_Date": pd.date_range("2020-01-01", periods=100, freq="M")
})

# --- Loop through columns ---
for col in df.columns:
    plt.figure(figsize=(6, 4))
    
    # NUMERIC columns → histogram
    if pd.api.types.is_numeric_dtype(df[col]):
        plt.hist(df[col], bins=15, color="skyblue", edgecolor="black")
        plt.title(f"Histogram of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
    
    # CATEGORICAL columns → bar chart
    elif pd.api.types.is_categorical_dtype(df[col]) or df[col].dtype == object:
        df[col].value_counts().plot(kind="bar", color="lightgreen", edgecolor="black")
        plt.title(f"Bar chart of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
    
    # DATETIME columns → line chart
    elif pd.api.types.is_datetime64_any_dtype(df[col]):
        plt.plot(df[col], range(len(df[col])), color="orange")
        plt.title(f"Line plot of {col} over index")
        plt.xlabel("Index")
        plt.ylabel(col)
    
    else:
        print(f"Skipping unsupported column type: {col}")
    
    plt.tight_layout()
    plt.show()
```

###  Code for estimation the value of **π (pi)** using a **Monte Carlo simulation**.

It works by randomly generating points within a unit square (from 0 to 1 on both axes) and checking how many of them fall inside a quarter circle of radius 1. The ratio of points inside the circle to the total number of points approximates the area ratio between the circle and the square. Multiplying that ratio by 4 gives an estimate of π.

**How it works:**
1. Import the `random` module to generate random numbers.
2. Set `n = 100000`, the number of random points to simulate.
3. For each point, generate random coordinates `(x, y)` in the range [0, 1].
4. Check if the point is inside the quarter circle using the condition `x**2 + y**2 <= 1`.
5. Count how many points (`inside`) meet that condition.
6. Estimate π using `pi_estimate = 4 * inside / n`.
7. Print the result with `print("Estimated π:", pi_estimate)`.

In essence, this simulates **throwing darts at a square dartboard** with a quarter circle drawn inside it — the fraction of darts landing inside the circle is used to approximate π.


### 🎲 How This Code Fits the Monte Carlo Definition

1. **Random sampling**  
   Each iteration generates random points `(x, y)` within a defined region — the unit square.  
   This randomness is the essence of the Monte Carlo approach.

2. **Statistical estimation**  
   You don’t compute π through geometry or calculus directly.  
   Instead, you estimate π by observing the ratio of random points that fall inside the quarter circle.

3. **Law of large numbers**  
   As you increase `n` (the number of random points), the estimate converges closer to the true value of π.  
   This statistical convergence is a hallmark of Monte Carlo methods.

4. **No deterministic formula**  
   You don’t use the actual formula for π.  
   You use probability and randomness to approximate it.


```python
import random
n = 100000
inside = 0
for _ in range(n):
    x, y = random.random(), random.random()
    if x**2 + y**2 <= 1:
        inside += 1
pi_estimate = 4 * inside / n
print("Estimated π:", pi_estimate)
```

# Algorithms 

```python
# Matrix Multiplication
A = [[1, 2, 3],
     [4, 5, 6]]

B = [[1, 2],
     [3, 4],
     [5, 6]]

rows_A = len(A)
cols_A = len(A[0])
cols_B = len(B[0])

C = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

for i in range(rows_A):
    for j in range(cols_B):
        for k in range(cols_A):
            C[i][j] += A[i][k] * B[k][j]

print("Result:\n", C)
```

```python
# Finding the Minimum Difference Pair
nums = [5, 3, 2, 4, 8]
nums.sort()

min_diff = float("inf")
pair = ()

for i in range(len(nums) - 1):
    diff = nums[i + 1] - nums[i]
    if diff < min_diff:
        min_diff = diff
        pair = (nums[i], nums[i + 1])

print("Closest pair:", pair, "Difference:", min_diff)
```
