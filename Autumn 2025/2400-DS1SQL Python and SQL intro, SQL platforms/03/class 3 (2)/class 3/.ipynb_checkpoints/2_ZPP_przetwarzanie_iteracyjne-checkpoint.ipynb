<h3>Rodzaje pętli</h3>
<p>Pętla to jedna podstawowych konstrukcji programowania. Umożliwia powtarzanie wybranego ciągu instrukcji określoną liczbę razy albo do spełnienia określonych warunk&oacute;w.</p>
<p>Możemy wyr&oacute;żnić dwa gł&oacute;wne typy pętli:</p>
<ul>
<li>iteracyjna (licznikowa)</li>
<li>repetycyjna (warunkowa)</li>
</ul>
<p>Pętla iteracyjna (lub <strong>pętla licznikowa</strong>) to rodzaj pętli, w ramach kt&oacute;rej następuje wykonanie określonej liczby iteracji (powt&oacute;rzeń). Do kontroli przebiegu wykonania pętli iteracyjnej stosuje się specjalną zmienną, w odniesieniu do kt&oacute;rej używa się określeń: <strong>zmienna sterująca</strong> (lub <strong>zmienna kontrolna</strong>, <strong>zmienna licznikowa</strong>). Przejście do kolejnej iteracji wiąże się ze zmianą wartości zmiennej sterującej o określoną wielkość i sprawdzeniem warunku, czy nowa wartość zmiennej sterującej znajduje się nadal w dopuszczalnym zakresie wartości, określonym dla tej zmiennej w ramach definicji pętli -- co jest r&oacute;wnoznaczne z powt&oacute;rzeniem pętli (wykonaniem kolejnej iteracji), czy też wartość ta znajduje się już poza zakresem -- co jest r&oacute;wnoznaczne z zakończeniem wykonywania pętli i przejściem do wykonania kolejnej instrukcji w programie. W Pythonie tego typu pętlę zdefiniujemy poleceniem `<span class="hljs-keyword">for</span> zmienna <span class="hljs-keyword">in</span> zbi&oacute;r_wartości:<span class="hljs-comment">&nbsp;POLECENIA</span>`.</p>
<p><br /><strong>Pętla repetycyjna</strong>&nbsp;(lub<strong> pętla warunkowa</strong>) to rodzaj pętli, w kt&oacute;rej wykonanie kolejnej iteracji uzależnione jest od pewnego, zdefiniowanego przez programistę warunku. Warunek zawarty w definiowanej pętli jest wyrażeniem, kt&oacute;re zwraca wartość typu logicznego. Przeważnie wyrażenie testujące sprawdzane jest na początku lub na końcu pętli. W zależności od tego czy wyrażenie jest prawdziwe, czy fałszywe, wykonywanie pętli jest kontynuowane, bądź przerywane. W R tego typu pętlę zdefiniujemy poleceniem `while()` albo `repeat()`.</p>

<h2>1. pętla iteracyjna -- polecenie for()</h2>

```python
# Tworzymy wektor z liczbami od 11 do 20 z użyciem pakietu numpy 
import numpy as np
wektor = np.arange(11, 21)

# Pętla, która wypisuje wektor 10 razy
for i in range(1,10):
    print(wektor)
```

Wyjaśnienie:
1. range(11, 21): Tworzy sekwencję liczb od 11 do 20 w Pythonie.
2. for i in range(10): Pętla, która wykonuje się 10 razy (od i = 0 do i = 9).
3. print(wektor): Wypisuje zawartość wektor w każdej iteracji pętli.

Powyższy kod należy odczytać następująco: dla zmiennej `i` o kolejnych wartościach od 1 do 10 wykonaj polecenia wewnątrz pętli (zaczyna się po : i jest wcięcie) -- za każdym razem wypisz obiekt `wektor`. Zauważmy, że polecenie wewnątrz pętli nie zależy od wartości zmiennej sterującej `i`. 


```python
for i in range(1,10):
    print(wektor[i])
```

```python
for i in range(0,10):
    print(wektor[i])
```

Dlaczego te dwa przykłady się różnią? - w Pythonie indexujemy od zera :)

Jako granice przedziałów można podać wyrażenia zwracające odpowiednie wartości:

```python
for i in range(5 - 4, len(wektor)):
    print(wektor[i])
```

Można też wskazać wybrane wartości:

```python
wektor = np.arange(11, 21)
for i in [2, 5, 10]:  
    if i <= len(wektor):  # Avoid index out of range error
        print(wektor[i - 1])  # Convert 1-based to 0-based indexing
    else:
        print(f"Index {i} out of bounds")
```

```python
for i in range(0, len(wektor), 3):  # Python uses zero-based indexing
    print(wektor[i])
```

Zmienna indeksowa nie musi być liczbą. Poza tym pętla może zawierać więcej poleceń.

```python
# Pętla for
for litera in ["s", "t", "u", "D", "i", "A"]:
    if litera == "s":
        litery = litera.upper()  # Jeśli litera to "s", zamieniamy ją na wielką
    else:
        litery = litery + litera.lower()  # Dla pozostałych liter dodajemy małe litery
    print(litery)
```

```python
# Zauważmy, że po zakończeniu pętli zmienna indeksowa z ostatnią wartością jest zapisana w przestrzeni roboczej.
print(litera)
```

Pętle można też w sobie zagnieżdżać. Stwórzmy przykładową macierz i wypełnijmy ją wartościami:


```python
import numpy as np

# Tworzenie macierzy 10x3 wypełnionej zerami
macierz = np.zeros((10, 3), dtype=int)

# Inicjalizacja liczby
liczba = 21

# Pętla do iteracji po wierszach i kolumnach
for i in range(macierz.shape[0]):  # Przechodzimy po wierszach
    for j in range(macierz.shape[1]):  # Przechodzimy po kolumnach
        macierz[i, j] = liczba
        liczba += 1  # Zwiększamy licznik o 1

# Wyświetlenie wynikowej macierzy
print(macierz)
```

<h2>2. pętla repetycyjna -- polecenie `while()`</h2>
<p>Pętlę repetycyjną możemy wykonać w R wykorzystując polecenie `while()`: `while( WARUNEK ): POLECENIA `. Należy to rozumieć jako: dop&oacute;ki WARUNEK jest spełniony wykonuj POLECENIA. Aby pętla `while()` była skończona, jedno z POLECEŃ musi wpływać na WARUNEK. W innym przypadku otrzymamy pętlę nieskończoną (o ile początkowo warunek był spełniony). W pętli `while()` warunek jest sprawdzany PRZED wykonaniem POLECEŃ. Jeśli więc warunek jest spełniony już w momencie wywołania pętli, nie wykona ona żadnej iteracji.</p>
<p>Na przykład:</p>

```python
# Pierwsza pętla

# Pierwsza pętla while i < 6: 
# Pętla będzie się wykonywać, dopóki zmienna i jest mniejsza niż 6. Wartość i jest zwiększana o 1 przy każdej iteracji.

i = 1

while i < 6:
    print(i)
    i += 1  # Inkrementujemy i o 1
```

```python
# Druga pętla

# Druga pętla while i < 6: Ta pętla się nie wykona, ponieważ początkowa wartość i wynosi już 6, więc warunek i < 6 jest fałszywy od samego początku.

i = 6

while i < 6:
    print(i)
    i += 1  # Inkrementujemy i o 1
```

```python
# Pętla while True w Pythonie jest używana do tworzenia nieskończonej pętli. Dopóki warunek w while jest zawsze True, pętla będzie się wykonywać w nieskończoność.

# while True:
#     print("Ewa")

count = 0

while True:
    print("Ewa")
    count += 1
    if count >= 5:
        break  # Przerwij pętlę po 5 iteracjach
```

dodakowy przykad while

```python
x = 4

while True:  # Nieskończona pętla
    print(x)
    x += 1  # Inkrementacja x o 1
    if x >= 5:  # Warunek przerwania pętli
        break
```

1. while True:: Tworzymy nieskończoną pętlę, która będzie działać do momentu, aż użyjemy break.
2. x += 1: Zwiększamy wartość zmiennej x o 1 w każdej iteracji.
3. if x >= 5:: Sprawdzamy, czy wartość x osiągnęła 5 lub więcej.
4. break: Kończymy pętlę, gdy x jest większe lub równe 5

<h2>3. polecenie `break`</h2>

Polecenia `break` możemy używać w pętlach `while` i `for`. Powoduje ono zawsze zakończenie wykonywania pętli i przejście do kolejnych poleceń. Podobnie można wykorzystać polecenie `next`, które kończy aktualnie wykonywaną iterację i przechodzi do wykonywania kolejnej.

```python
for i in range(-10, 11):  # Iteracja od -10 do 10 (włącznie)
    if -5 <= i <= 5:  # Jeśli i mieści się w przedziale [-5, 5], przerwij pętlę
        break
    print(i)
```

Wyjaśnienie:
* range(-10, 11): Tworzy zakres liczb od -10 do 10 włącznie.
* if -5 <= i <= 5:: Sprawdzamy, czy wartość i mieści się w przedziale od -5 do 5. Jeśli tak, pętla zostaje przerwana za pomocą break.
* print(i): Jeśli warunek nie jest spełniony, drukujemy wartość i.
* pętla zatrzymuje się, gdy i osiągnie -5.

<h2>4. polecenie `continue`</h2>

```python
for i in range(-10, 11):  # Iteracja od -10 do 10 (włącznie)
    if -5 <= i <= 5:  # Jeśli i mieści się w przedziale [-5, 5], przejdź do następnej iteracji
        continue
    print(i)
```

Wyjaśnienie:
* continue: Zamiast przerywać pętlę, przechodzi do następnej iteracji, pomijając wykonanie print(i) dla wartości w przedziale -5 <= i <= 5.
* print(i): Drukuje wartość i tylko wtedy, gdy nie mieści się w przedziale [-5, 5].

Możemy uzyć obu poleceń w jednej pętli, różne typy pętli możemy w sobie zagnieżdżać.

W poniższym przykładzie dla liczb 2-10 sprawdźmy:

* wartość której potęgi danej liczby po raz pierwszy przekracza 1000 (`1e3`),
* wartość której potęgi danej liczby po raz pierwszy przekracza 1000000 (`1e6`).

```python
# Pętla for iterująca po liczbach od 2 do 10
for liczba in range(2, 11):
    potega = 1
    
    # Pętla while (odpowiednik repeat w R)
    while True:
        # Czy kolejna potęga po raz pierwszy przekracza 1000
        if liczba ** potega > 1e3 and liczba ** (potega - 1) <= 1e3:
            print(f"{potega} potęga liczby {liczba} po pierwszy przekracza 1000")
            potega += 1
            continue  # Kontynuujemy kolejną iterację pętli while
        
        # Czy kolejna potęga przekracza 1000000
        if liczba ** potega > 1e6:
            print(f"{potega} potęga liczby {liczba} po pierwszy przekracza 1000000")
            break  # Kończymy pętlę while
        
        # Zwiększamy potęgę o 1
        potega += 1
```

<h3><strong>Wyjaśnienie:</strong></h3>
<ol>
<li>
<p><strong>Pętla <code>for liczba in range(2, 11)</code></strong>: Tworzymy pętlę, kt&oacute;ra przechodzi przez liczby od 2 do 10 (w Pythonie <strong><code>range(2, 11)</code></strong> tworzy zakres od 2 do 10 włącznie).</p>
</li>
<li>
<p><strong>Pętla <code>while True</code></strong>: Używamy nieskończonej pętli <strong><code>while</code></strong>, ponieważ chcemy ręcznie przerwać ją na podstawie warunk&oacute;w (odpowiednik pętli <strong><code>repeat</code></strong> w R).</p>
</li>
<li>
<p><strong><code>if liczba ** potega &gt; 1e3 and liczba ** (potega - 1) &lt;= 1e3:</code></strong>: Sprawdzamy, czy kolejna potęga liczby przekracza 1000 po raz pierwszy. Jeśli tak, drukujemy komunikat i przechodzimy do kolejnej potęgi (zwiększamy wartość potęgi i używamy <strong><code>continue</code></strong> do przejścia do następnej iteracji pętli <strong><code>while</code></strong>).</p>
</li>
<li>
<p><strong><code>if liczba ** potega &gt; 1e6:</code></strong>: Sprawdzamy, czy potęga przekracza 1 000 000. Jeśli tak, drukujemy komunikat i kończymy pętlę <strong><code>while</code></strong> za pomocą <strong><code>break</code></strong>.</p>
</li>
<li>
<p><strong>Zwiększanie potęgi</strong>: Na końcu każdej iteracji pętli zwiększamy potęgę o 1.</p>
