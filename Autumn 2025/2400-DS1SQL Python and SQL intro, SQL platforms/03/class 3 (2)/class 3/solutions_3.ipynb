```python
# Exercise 1
# ----------------------------------------
# Task:
# Using an appropriate conditional statement and/or loop, check for each number 
# in a vector of natural numbers whether it is even or odd.
# The % operator (modulo) may be useful.

numbers = list(range(1, 21))  # Example vector of natural numbers from 1 to 20

for num in numbers:
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
# Exercise 2
# ----------------------------------------
# Task:
# Write a loop that iterates through a list of all letters (both lowercase and uppercase) 
# and displays only the vowels.

import string

letters = string.ascii_lowercase   # 'abcdefghijklmnopqrstuvwxyz'
LETTERS = string.ascii_uppercase   # 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

vowels = "aeiouAEIOU"

print("Vowels in lowercase letters:")
for letter in letters:
    if letter in vowels:
        print(letter, end=" ")

print("\n\nVowels in uppercase letters:")
for letter in LETTERS:
    if letter in vowels:
        print(letter, end=" ")
```

```python
# Exercise 3
# ----------------------------------------
# Task:
# Using an appropriate loop, print successive natural numbers,
# but only until the square of the number exceeds 1000.

num = 1
while num ** 2 <= 1000:
    print(num)
    num += 1

print(f"Stopped because {num}^2 = {num**2} > 1000")
```
